```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import scipy.stats

cars_df = pd.read_csv('data/Cars93.csv')
```

```python
cars_df[['Weight','MPG.highway']].corr()
```

```python
fig = plt.figure(figsize = (10,8))
plt.title("Weight of Vehicles vs. MPG", fontsize = 20)
ax = sns.scatterplot(data = cars_df,
                     x = 'Weight',
                     y = 'MPG.highway',
                     hue = 'Weight',
                     palette = 'magma', 
                     s = 100,
                     alpha = .8, 
                     edgecolor = 'black', 
                     linewidth = 1,
                     legend = False)

ax.set_xlabel('Weight', fontsize = 14)
ax.set_ylabel('MPG (Highway)', fontsize = 14)
ax.tick_params(size = 10)
ax.set_yticklabels(ax.get_yticks(), size = 13)
ax.set_xticklabels(ax.get_xticks(), size = 13)
sns.despine()
plt.show()
```

```python
x = cars_df['Weight'].to_numpy()
y = cars_df['MPG.highway'].to_numpy()

slope, intercept, r, p, stderr = scipy.stats.linregress(x,y)
line = f'Slope: {slope:.2f}\nIntercept: {intercept:.2f}\nCorrelation coefficient: r={r:.2f}'

fig = plt.figure(figsize = (10,8))
plt.title("Weight of Vehicles vs. MPG", fontsize = 20)
ax = sns.scatterplot(data = cars_df,
                     x = 'Weight',
                     y = 'MPG.highway',
                     hue = 'Weight',
                     palette = 'magma', 
                     s = 100,
                     alpha = .8, 
                     edgecolor = 'black', 
                     linewidth = 1,
                     legend = False)

ax = sns.lineplot(x,
                  intercept + slope * x, 
                  label = line,
                  color = 'blue',
                  linewidth = 1.2)

ax.set_xlabel('Weight', fontsize = 14)
ax.set_ylabel('MPG (Highway)', fontsize = 14)
ax.tick_params(size = 10)
ax.set_yticklabels(ax.get_yticks(), size = 13)
ax.set_xticklabels(ax.get_xticks(), size = 13)
sns.despine()
plt.show()
```
